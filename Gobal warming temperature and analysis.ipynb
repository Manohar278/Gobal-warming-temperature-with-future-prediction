### Outlier

```python
import matplotlib.pyplot as plt
```

```python
data_long=data.melt(
    id_vars=['Country'],
    var_name='Year',
    value_name='Temperature'
)
```

```python
data_long
```

```python
data_long['Year']=data_long['Year'].str.replace("F","").astype(int)
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(data_long['Temperature'])
plt.title('Global Temperature Outliers')
plt.ylabel('Temperature Anomaly')
plt.show()
```

### Alternative to melt

```python
data_stack = data.set_index("Country").stack().reset_index()
data_stack.columns = ["Country", "Year", "Temperature"]

data_stack["Year"] = data_stack["Year"].str.replace("F", "").astype(int)
```

### Model Building

```python
x=data_long[['Year']]
y=data_long[['Temperature']]
```

### Train the linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
print("slope(Warming rate per year):",model.coef_[0])
print("Intercept:",model.intercept_)
```

```python
yearly_avg=data_long.groupby("Year")["Temperature"].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.plot(yearly_avg.index,yearly_avg.values)
plt.xlabel("Year")
plt.ylabel('Temperature')
plt.title("Global Temperature Trend")
plt.show()
```

```python
next_year=[[2023]]
predicted_temp=model.predict(next_year)
predicted_temp[0]
```

A line and bar chart of yearly average temperature anomalies showed a clear increasing trend. Linear regression further confirmed a positive relationship between year and temperature, indicating global warming.
